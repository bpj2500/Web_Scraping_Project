```python
#loading in the datasets 
import glob 
import pandas as pd 

path = r'C:\Users\bpj25\OneDrive\Desktop\Introduction_to_Scrapy\realtor' 

allfiles = glob.glob(path + "/*.csv")

df = pd.concat(map(pd.read_csv, allfiles))


```

```python
#cleaning up the data 
#Frirst dropping all the rows with no data at all
df = df.dropna(axis = 0, how='all') 


#check
df.head()
```

```python
#Adding a zipcode column
df['zipcode'] = df['area'].str.extract(r'(\d+)') 
df.head()
```

```python
#cleaning the prices column of symbols
prices = df['price'].str.replace('$', '').str.replace(',','') 

df['price'] = [int(x) for x in prices] 

df.head()
```

```python
#getting rid of the plus symbols in baths column

baths = df['baths'] 
baths = [str(x) for x in baths] 
baths =[float(x.replace('+', '')) for x in baths]
df['baths'] = baths 
df.head()
```

```python
#calculating the down payment for buying the estate  

high_payment = [0.2 * x for x in df['price']] 
low_payment = [0.05 * x for x in df['price']] 

df['high_payment'] = high_payment 
df['low_payment'] = low_payment 

df.head()
```

```python
#calculating monthly mortage for a high down payment
x = df['price']  
y = df['high_payment']

high_monthly = list(round((x - y)/360)) 
df['high_monthly'] = high_monthly 

df.head()
```

```python
#calculcating the monthly mortage for a low down payment 
x = df['price']  
y = df['low_payment']

low_monthly = list(round((x - y)/360)) 
df['low_monthly'] = low_monthly 

df.head()
```

```python
#both cells are calculating the annuanl income reccomended to afford a mortage with a high down payment
annual_income = [round((12*x*100)/30) for x in df['high_monthly']]
```

```python
df['annual_income_high'] = annual_income 

df.head()
```

```python
#calculating the annuanl income reccomended to afford a mortage with a low down payment
annual_income = [round((12*x*100)/30) for x in df['low_monthly']] 

df['annual_income_low'] = annual_income 

df.head()
```

```python
#writing to csv
df.to_csv(r'C:\Users\bpj25\OneDrive\Desktop\Introduction_to_Scrapy\realtor\real_estate_data.csv')
```
